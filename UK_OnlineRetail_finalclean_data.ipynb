```python
df_clean = df_clean[~df_clean['InvoiceNo'].astype(str).str.startswith('C')]                      #cancelled ordered removed
df_clean.shape
```

```python
# ============================
# CLEAN THE DESCRIPTION COLUMN
# ============================

# 1. Remove leading and trailing spaces
# Example:
# "  PARTY BUNTING  " --> "PARTY BUNTING"
df_clean["Description"] = df_clean["Description"].str.strip()


# 2. Convert all descriptions to UPPERCASE
# Makes the text consistent.
# Example:
# "Party Bunting" --> "PARTY BUNTING"
df_clean["Description"] = df_clean["Description"].str.upper()


# 3. Replace multiple spaces with a single space
# Example:
# "BOX   OF   VINTAGE   JIGSAW BLOCKS"
# -->
# "BOX OF VINTAGE JIGSAW BLOCKS"
df_clean["Description"] = df_clean["Description"].str.replace(r"\s+", " ", regex=True)


# 4. Remove unwanted symbols only from the beginning and end
# (Keeps symbols inside genuine product names)
# Example:
# "*** PARTY BUNTING ***"
# -->
# "PARTY BUNTING"
df_clean["Description"] = df_clean["Description"].str.strip("*#.,!?;:")


# 5. Remove extra quotation marks if present
# Example:
# '"PARTY BUNTING"' --> PARTY BUNTING
df_clean["Description"] = df_clean["Description"].str.replace('"', "", regex=False)

# ============================
# STANDARDIZE PUNCTUATION
# ============================

# Remove commas from descriptions
df_clean["Description"] = df_clean["Description"].str.replace(",", "", regex=False)

# Remove period (.) only if it appears at the end of the description
df_clean["Description"] = df_clean["Description"].str.replace(r"\.$", "", regex=True)

# Remove extra spaces created after cleaning
df_clean["Description"] = df_clean["Description"].str.replace(r"\s+", " ", regex=True).str.strip()



                                                  # Check the number of unique descriptions after cleaning
df_clean["Description"].nunique()
```
